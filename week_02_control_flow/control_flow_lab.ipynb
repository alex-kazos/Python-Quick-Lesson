## Booleans and comparisons

```python
guess = 7
secret_number = 10

print(guess == secret_number)
print(guess < secret_number)
print(guess > secret_number)
```

## Decisions

```python
if guess == secret_number:
    print('Correct')
elif guess < secret_number:
    print('Too low')
else:
    print('Too high')
```

## Loops and lists

```python
guesses = [3, 8, 10]

for current_guess in guesses:
    print(f'Checking {current_guess}')
    if current_guess == secret_number:
        print('Found it')
```

## Mini challenge

Create a list of quiz answers. Count how many are correct.

```python
answers = ['python', 'git', 'pandas']
correct_answers = ['python', 'github', 'pandas']
score = 0

for index, answer in enumerate(answers):
    if answer == correct_answers[index]:
        score += 1

print(f'Score: {score}/{len(correct_answers)}')
```
