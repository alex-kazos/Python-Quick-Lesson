## First function

```python
def greet(name):
    return f'Hello, {name}'

greet('Alex')
```

## Parameters and return values

```python
def calculate_expense_total(expenses):
    total = 0
    for expense in expenses:
        total += expense['amount']
    return total

expenses = [
    {'name': 'Coffee', 'amount': 3.50, 'category': 'food'},
    {'name': 'Bus', 'amount': 1.20, 'category': 'transport'},
    {'name': 'Game', 'amount': 19.99, 'category': 'gaming'},
]

calculate_expense_total(expenses)
```

## Category totals

```python
def total_by_category(expenses, category):
    total = 0
    for expense in expenses:
        if expense['category'] == category:
            total += expense['amount']
    return total

total_by_category(expenses, 'food')
```
